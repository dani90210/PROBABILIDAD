```python
import statistics as stats

grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)

# Promedio con sum()
promedio = sum(grupoa) / len(grupoa)
print("Promedio con sum:", promedio)

# Promedio con statistics.mean()
promedio2 = stats.mean(grupoa)
print("Promedio con mean:", promedio2)

# Mediana
mediana = stats.median(grupoa)
print("Mediana:", mediana)

# Multimoda
moda2 = stats.multimode(grupoa)
print("Moda(s):", moda2)

# Lista de tiempos
tiempos100mts = [15.10, 17.20, 14.69, 13.27, 22.15, 18.71, 19.5, 20, 15.10, 17.20]
print(tiempos100mts)

# Promedio
print("El tiempo promedio de los corredores:", stats.mean(tiempos100mts))

# Mediana
mediana = stats.median(tiempos100mts)
print("La mediana de los tiempos de los corredores es:", mediana)

# Moda
moda = stats.mode(tiempos100mts)
print("La moda de los tiempos es:", moda)

# Multimodas
multimodas = stats.multimode(tiempos100mts)
print("Las modas de los tiempos son:", multimodas)
```

```python
from statistics import*
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(0,100),2)
  fabrica1.append(numero_aleatorio)
  print(fabrica1)
```

```python
fabrica2= []
for i in range(100):
  numero_aleatorio = round(random.uniform(40,93),2)
  fabrica2.append(numero_aleatorio)
  print(fabrica2)

```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("producciond de la fabrica1")
```

```python

plt.hist(fabrica2)
plt.title("producci√≥n diaria de la fabrica 2")
```

```python
media1 = mean (fabrica1)
mediana1 = median (fabrica1)
moda1 = mode (fabrica1)

print ("valores de la fabrica1")
print ("la media es =", media1)
print ("la mediana es =", mediana1)
print ("la moda es=", moda1)
```

```python
media2 = mean (fabrica2)
mediana2 = median (fabrica2)
moda2 = mode (fabrica2)

print("valores de la fabrica2")
print ("la media es =", media2)
print ("la mediana es =", mediana2)
print("la moda es=", moda2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize=(10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python

plt.hist (fabrica1, bins=10, alpha=0.2, color='b',  edgecolor = "black")
plt.axvline (media1, color = 'b', linestyle ='dashed', linewidth = 2, label = "Media1")
plt.axvline (mediana1, color = 'g', linestyle ='dashed', linewidth = 2, label = "Mediana1")
plt.axvline (moda1, color = 'r', linestyle ='dashed', linewidth = 2, label = "Moda1")
plt.xlabel ('Valor')
plt.ylabel ('Frecuencia')
plt.title ("Produccion diaria de la fabrica1")
plt.legend()
plt.show()
```
